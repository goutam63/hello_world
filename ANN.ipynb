```
x=data.iloc[:,:-1]
```

```
y=data.iloc[:,-1]
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=42)
```

keras frame work

```
from keras import Sequential # seuential model our Ann is basically Sequential model
```

```
from keras.layers import Dense
```

```
import keras
```

```
model=Sequential() # initialize model
```

## adding layers

```
# adding layers
model.add(Dense(units = 8, input_dim = 8)) # 
model.add(Dense(units = 12, activation = 'relu'))
model.add(Dense(units = 1, activation = 'sigmoid'))
```

```
model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

## train the model

```
history=model.fit(x,y,epochs=500,validation_data=(x_test,y_test), batch_size=10,verbose=1)
```

epochs= it is itaration, 
batch_size= it is number size of batch is model,
sgd=secastic grediant degredation,

```
print(history.history.keys())
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model acuuracy')
plt.xlabel('ephoch')
plt.ylabel('accuracy')
plt.legend(['train','test'],loc='upper left')
plt.show()
```
